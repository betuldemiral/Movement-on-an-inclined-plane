Lagragian Method:

$$
L = T - V
$$

T : Kinetic Energy of the system \\
V : Potential Energy of the system \\

Equation of Lagrange:

$$
\frac{d}{dt}(\frac{əL}{ə q^ ̇ }) - \frac{əL}{əq} = Q_i
$$

In this problem there is no force acting from outside so;

$$
Q_i = 0
$$

Lagrange's equation becomes this
$$
\frac{d}{dt}(\frac{əL}{ə q^ ̇ }) - \frac{əL}{əq} = 0
$$


Moment of inertia for disc : $$ I =  \frac{1}{2}mr^2 $$


In this case we have 2 different kinetic energy(T and T(Rotational))

$$ T = \frac{1}{2}m({x^̇ }^2+ {y^ ̇ }^2 + {z^ ̇ }^2)$$  we will use only x term so;

$$ T = \frac{1}{2}m{x^̇ }^2$$

Rotational Kinetic Energy :
$$
T_R = \frac{1}{2}Iω^2
$$

$$
ω^2 = θ^̇
$$

$$
T_R = \frac{1}{2}I{θ^̇ }^2
$$


As you can see, kinetic energy includes $x^̇ $ and $θ^̇ $ variables. For the Lagrange solution, we convert $θ^̇ $ value to $x^̇$ with the following transformation.

circ.PNG


$$
x = rθ \\
x^̇  = r θ ^̇
$$

$$
 θ ^̇  = \frac{x^̇ }{r} \\
(θ ^̇ )^2 = {(\frac{x^̇ }{r})}^2
$$

$$
T_R = \frac{1}{2}I{θ^̇ }^2 = \frac{1}{2}I{(\frac{x^̇ }{r})}^2
$$

If we put the moment of the inertia to the $T_R$

$$
T_R = \frac{1}{2}(\frac{1}{2}mr^2){(\frac{x^̇ }{r})}^2
$$

$$
T_R = \frac{1}{4}(mr^2)(\frac{{x^̇ }^2}{r^2})
$$

$$
T_R = \frac{1}{4}m{x^̇ }^2
$$

The Total Kinetic Energy will be;

$$
T_{total} = T + T_R
$$

$$
T_{total} = \frac{1}{2}m{x^̇ }^2 + \frac{1}{4}m{x^̇ }^2
$$

$$
T_{total} = \frac{3}{4}m{x^̇ }^2
$$

Potential Energy of the system:

$$
V = mgh \\
h = l - xsinα \\
V = mg(l-xsinα)
$$


We found the total Kinetic Energy and Potential Energy for the system so we can write the Lagrangian(L).

$$
L = T - V
$$

$$
L = \frac{3}{4}m{x^̇ }^2 - mg(l-xsinα)
$$

$$
L = \frac{3}{4}m{x^̇ }^2 - mgl + mgxsinα
$$

Calculation of Lagrange:

$$
\frac{d}{dt}(\frac{əL}{ə x^ ̇ }) - \frac{əL}{əx} = 0
$$

firstly find the derivatives

$$
\frac{əL}{ə x^ ̇ } = \frac{ə(\frac{3}{4}m{x^̇ }^2 - mgl + mgxsinα)}{ə x^ ̇ } = \frac{3}{2}mx^{̇ }
$$

\\

$$
\frac{d}{dt}(\frac{əL}{ə x^ ̇ }) = \frac{d}{dt}(\frac{3}{2}mx^{̇ }) = \frac{3}{2}mx^{̈ }
$$

\\

$$
\frac{əL}{əx} = \frac{ə(\frac{3}{4}m{x^̇ }^2 - mgl + mgxsinα)}{əx} = mgsinα
$$
 \\

 $$
 \frac{3}{2}mx^{̈ } - mgsinα = 0
 $$
 \\

 $$
 \frac{3}{2}mx^{̈ } = mgsinα
 $$

 \\

$$
\frac{3}{2}x^{̈ } = gsinα
$$

 \\

$$
x^{̈ } = \frac{2}{3}gsinα
$$

```python
from math import sin, pi, tan
import matplotlib.pyplot as mpl_p
import numpy as np
from IPython.display import display, clear_output
import time

g=9.81

class InclinedPlane:
    def __init__(self, length, angle):
        self.length = length
        self.angle = np.radians(angle)

    def height_at_position(self, position):
        return position * np.tan(self.angle)

    def angle_to_rad(self):
        rad_angle = pi * self.angle / 180
        return rad_angle

class SlidingDisc:
    def __init__(self, mass, radius, initial_position, initial_velocity,dt):
        self.mass = mass
        self.radius = radius
        self.position = initial_position
        self.velocity = initial_velocity
        self.dt = dt

    def update_position(self):
        self.position += self.velocity * self.dt

class Simulation:
    def __init__(self, inclined_plane, sliding_disc,num_step):
        self.inclined_plane = inclined_plane
        self.sliding_disc = sliding_disc
        self.num_step = num_step

    def simulate(self):
      positions = [self.sliding_disc.position]

      fig, ax = mpl_p.subplots()
      mpl_p.xlabel("Position (m)")
      mpl_p.ylabel("Height (m)")
      mpl_p.title("Inclined Plane Simulation")
      mpl_p.grid(True)

      disc_point, = ax.plot([], [], 'ro')
      inclined_line, = ax.plot([], [], 'b--')


      disc_point.set_data(self.sliding_disc.position, 0)
      inclined_line.set_data([0, self.inclined_plane.length],
       [0, self.inclined_plane.height_at_position(self.inclined_plane.length)])


      for _ in range(self.num_step):
        height = self.inclined_plane.height_at_position(self.sliding_disc.position)
        self.sliding_disc.update_position()

        disc_point.set_data(self.sliding_disc.position, height)
        inclined_line.set_data([0, self.inclined_plane.length],
         [0, self.inclined_plane.height_at_position(self.inclined_plane.length)])

        mpl_p.xlim(0, self.inclined_plane.length)
        mpl_p.ylim(0, max(self.inclined_plane.height_at_position(self.inclined_plane.length), 1))
        display(fig)
        clear_output(wait=True)
        time.sleep(0.001)

        positions.append(self.sliding_disc.position)

    mpl_p.show()




class LagrangeProblemSol(InclinedPlane , SlidingDisc):
    def __init__(self, length,angle,mass,radius, initial_position , initial_velocity, dt,t_0,n):
        InclinedPlane.__init__(self, length=length, angle=angle)
        SlidingDisc.__init__(self, mass=mass, radius=radius, initial_position=initial_position ,
                             initial_velocity=initial_velocity,dt=dt)
        self.t_0 = t_0
        self.n = n

    def euler_lagrange_sol(self):
        times = [self.t_0]
        Xs = [self.position]
        V_Xs = [self.velocity]

        for _ in range(self.n):
            x = self.position + self.velocity * self.dt
            t = self.t_0 + self.dt
            V_x = self.velocity - ((2 * g * sin(self.angle_to_rad())) / 3) * self.dt

            times.append(t)
            Xs.append(x)
            V_Xs.append(V_x)

            self.t_0, self.position, self.velocity = t, x, V_x
        return times, Xs, V_Xs


    def plot(self):
      mpl_p.figure(figsize=(12, 6))
      mpl_p.subplot(1, 2, 1)
      mpl_p.plot(times, xs, label="x")
      mpl_p.xlabel("Time (s)")
      mpl_p.ylabel("Position (m)")
      mpl_p.legend()

      mpl_p.subplot(1, 2, 2)
      mpl_p.plot(times, v_xs, label="v_x")
      mpl_p.xlabel("Time (s)")
      mpl_p.ylabel("Velocity (m/s)")
      mpl_p.legend()

      mpl_p.tight_layout()
      mpl_p.show()






inclined_plane = InclinedPlane( length=10, angle=30)
sliding_disc = SlidingDisc(mass=1, radius=0.1, initial_position=10, initial_velocity=-1,dt=0.05)

simulation = Simulation(inclined_plane, sliding_disc,num_step=199)
simulation.simulate()


lagrange_solver = LagrangeProblemSol(length=10, angle=40, mass=10, radius=5.0,
                                     initial_position=0.0,initial_velocity = 0.0,dt=0.5,t_0=0,n=10)
times, xs, v_xs = lagrange_solver.euler_lagrange_sol()
lagrange_solver.plot()
```
